````python
from ultralytics import YOLO
````

````python
MODEL_NAME = ""
IMAGE_SIZE = 640
BATCH = 64
CONF = 0.25
IOU = 0.6

def print_metrics(metrics):
    print(f"""
**model {MODEL_NAME}**
---
Params:
```
image_size :  {IMAGE_SIZE}
batch      :  {BATCH}
conf_tresh :  {CONF}
iou_tresh  :  {IOU}
```

Results:
```
precision  : {metrics.box.p[0]:.4f}
recall     : {metrics.box.r[0]:.4f}
f1 score   : {metrics.box.f1[0]:.4f}
ap50       : {metrics.box.ap50[0]:.4f}
ap         : {metrics.box.ap[0]:.4f}
mp         : {metrics.box.mp:.4f}
mr         : {metrics.box.mr:.4f}
map50      : {metrics.box.map50:.4f}
map75      : {metrics.box.map75:.4f}
```

""")
````

````python
MODEL_NAME = "yolo11n.pt"
print_metrics(YOLO(MODEL_NAME).val(data="test_dataset.yml", imgsz=IMAGE_SIZE, batch=BATCH, conf=CONF, iou=IOU))
````

````python
MODEL_NAME = "yolo11s.pt"
print_metrics(YOLO(MODEL_NAME).val(data="test_dataset.yml", imgsz=IMAGE_SIZE, batch=BATCH, conf=CONF, iou=IOU))
````
